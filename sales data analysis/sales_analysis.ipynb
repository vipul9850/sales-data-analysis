```python
df['ORDERDATE'] = pd.to_datetime(df['ORDERDATE'])

```

```python
df=df.drop(columns=['ADDRESSLINE2'])
```

```python
df = df.rename(columns={'ADDRESSLINE2': 'ADDRESSLINE'})
```

```python
df['STATE']=df['STATE'].fillna('unknown')
```

```python
df['TERRITORY']=df['TERRITORY'].fillna('Not Assigned')
```

```python
df['POSTALCODE']=df['POSTALCODE'].fillna('NA')
```

```python
df.isnull().sum()
df.info()
```

```python
product_sales = df.groupby("PRODUCTLINE")["SALES"].sum().sort_values(ascending=False)
print(product_sales)
```

```python
monthly_sales=df.groupby("MONTH_ID")["SALES"].sum()
print(monthly_sales)
```

```python
yearly_sales=df.groupby("YEAR_ID")["SALES"].sum()
print(yearly_sales)
```

```python
country_sales=df.groupby("COUNTRY")["SALES"].sum().sort_values(ascending=False)
print(country_sales.head(10))
```

```python
deal_sales=df.groupby("DEALSIZE")["SALES"].sum()
print(deal_sales)
```

```python
monthly_sales.plot(kind="line",marker="o")
plt.title("Monthly Sales Trend")
plt.xlabel("Month")
plt.ylabel("Total Sales")
plt.show()
```

```python
df["ORDERDATE"] = df["ORDERDATE"].dt.date
```
